```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
uploaded  = files.upload()
```

```python
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
cv2_imshow(img)
```

```python
gyay = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gyay)
```

```python
resized =cv2.resize(gyay, (150,150))
cv2_imshow(resized)
```

```python
resized =cv2.resize(img, (150,150))
cv2_imshow(resized)
```

```python
a = [img,gyay]
for i in a:
  cv2_imshow(i)
resized =cv2.resize(img, (150,150))
cv2_imshow(resized)
resized =cv2.resize(gyay, (150,150))
cv2_imshow(resized)
```

```python
annotated = img.copy()
cv2.rectangle(annotated, (60, 40), (250, 250), (255, 255, 0),2)
cv2.putText(annotated,'Sahjad',(70,60),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0),2)
cv2_imshow(annotated)
```

```python
cv2.imwrite('output_image.jpg',annotated)
print("Images saved as output_image.jpg")
```

```python
# Example No.2
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

```python
uploaded = files.upload()
filename =list(uploaded.keys())[0]
```

```python
img = load_img(filename,target_size=(224,224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array,axis=0)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=(0.5,1.5),
    fill_mode='nearest'
)
i = 0
for batch in datagen.flow(img_array,batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8'))
  plt.axis('off')
  plt.show()
  i+=1
  if i>5:
    break
```
